# Plotly Scatter Plot 

```python
import plotly
import plotly.plotly as plt
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import json
```

```python
data = pd.read_csv('masterPatient.csv')
data = data.sort_values(by=['SUBJECT_ID'])
data = data.reset_index(drop=True)
```

```python
with open('num_visits_lib.txt') as f:
    Num_Visits_Lib = json.load(f)
```

```python
v2 = []
v3 = [] 
v4 = []
v5 = []
v6 = []
v7 = []
v8 = []
for key,val in Num_Visits_Lib.items():  
    if(val == 2):
        v2.append(key)  
    elif(val == 3):
        v3.append(key)
    elif(val == 4):
        v4.append(key)
    elif(val == 5):
        v5.append(key)
    elif(val == 6):
        v6.append(key)
    elif(val == 7):
        v7.append(key)
    elif(val == 8):
        v8.append(key)
```

```python
x_data = []
y_data = []
color = []
for index,row in data.iterrows():
    if(str(row['SUBJECT_ID']) in v2):
        x_data.append(row['visit_num'])
        y_data.append(row['SUBJECT_ID'])  
        color.append(row['DiagnosesColor'])
```

```python
plotly.offline.init_notebook_mode(connected=True)

plotly.offline.iplot({
    "data": [go.Scatter(x=x_data, y=y_data, mode='markers',marker=dict(size=2,color=color))],
    "layout": go.Layout(title="Patients Diagnosed with DM2")
})
```
