# Scatterplot of Multiple Visit DM2 Patients

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_csv('masterPatient.csv')
data = data.sort_values(by=['SUBJECT_ID'])
data = data.reset_index(drop=True)
```

Example of data that is being represented:

```python
data.head()
```

```python
with open('num_visits_lib.txt') as f:
    Num_Visits_Lib = json.load(f)
```

```python
v2 = []
v3 = [] 
v4 = []
v5 = []
v6 = []
v7 = []
v8 = []
for key,val in Num_Visits_Lib.items():  
    if(val == 2):
        v2.append(key)  
    elif(val == 3):
        v3.append(key)
    elif(val == 4):
        v4.append(key)
    elif(val == 5):
        v5.append(key)
    elif(val == 6):
        v6.append(key)
    elif(val == 7):
        v7.append(key)
    elif(val == 8):
        v8.append(key)
```

Every patient represented on the below graph has at least one time not beeen diagnosed with DM2. Therefore, if during the first visit the patient was diagnosed with DM2, then the the second visit the patient could not have been diagnosed with DM2.

What would be expected in the graph below is that the left side (first visit) would be green and that the right side (second visit) would be red. However, this is not the case and instead what is seen is a rather random looking mixture. 

Following the first graph, are patients who have 3 vistits, 4 visits, etc. (up to 8 visits)

```python
fig = plt.figure(figsize=(5,30))
ax = fig.add_subplot(1,1,1)
for index,row in data.iterrows():
    if(str(row['SUBJECT_ID']) in v2):
        ax.scatter(row['visit_num'],row['SUBJECT_ID'],alpha=0.8,c=row['DiagnosesColor'],edgecolors='none', s=10)
plt.title('Patients Diagnosed with DM2')
plt.ylabel('Patients')
plt.xlabel('Visit (in order)')
plt.grid(axis='y')
#plt.legend(loc=2)
plt.show()
```
